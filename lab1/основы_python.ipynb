#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def countVowels(str):
    str = str.lower()
    c = 0
    for i in str:
        if i in 'aeiou':
            c+= 1
    return c

def test_countVowels():
    assert countVowels('hello') == 2

    assert countVowels('bcdfg') == 0

    assert countVowels('aeiou') == 5

    assert countVowels('') == 0
    
    assert countVowels('AEIOU') == 5  
    
    print("Все тесты пройдены")


test_countVowels()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def checkUnicElements(str):
    collection = set()
    str = str.lower()
    for i in str:
        if i in collection:
            return False
        collection.add(i)
    return True


def test_checkUnicElements():
    assert checkUnicElements('abcde') == True

    assert checkUnicElements('aabcde') == False

    assert checkUnicElements('') == True

    assert checkUnicElements('a') == True

    assert checkUnicElements('aA') == False  
   
    assert checkUnicElements('123!@#') == True
    
    print("Все тесты пройдены!")


test_checkUnicElements()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_ones_in_binary(num):
    c = 0
    while num > 0:
        if num%2 == 1:
            c+= 1
        num = num//2
    return c 

def test_count_ones_in_binary():
    assert count_ones_in_binary(5) == 2      
    assert count_ones_in_binary(15) == 4     

    assert count_ones_in_binary(1) == 1      
    assert count_ones_in_binary(0) == 0 

    assert count_ones_in_binary(255) == 8
    assert count_ones_in_binary(1024) == 1
    
    print("Все тесты пройдены!")
    
test_count_ones_in_binary()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    if n < 10:
        return 0

    count = 0
    current = n
    while current >= 10:
        
        result  = 1
        for i in str(current):
            result  *= int(i)
        current = result 
        count += 1
    
    return count

def test_magic():
    assert magic(39) == 3
    assert magic(4) == 0
    assert magic(999) == 4
    assert magic(10) == 1 
    assert magic(123) == 1  
    print("Все тесты пройдены!")

test_magic()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import math

def mse(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("length")
    if not v1:
        raise ValueError("empty")
    
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(v1, v2)) / len(v1))

def test_mse():
    assert mse([1, 2, 3], [1, 2, 3]) == 0.0
    assert mse([2, 3, 4], [1, 2, 3]) == 1.0
    assert mse([0, 0], [1, 1]) == 1.0
    assert mse([-1, -2], [1, 2]) == math.sqrt(20/2)

    assert mse([5], [3]) == 2.0

    assert mse([100, 200], [110, 190]) == math.sqrt(200/2)
 
    print("Все тесты прошли")


test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def razlo(n):
    if n< 0 or (n% 1 != 0):
        return False
    
    if n < 2:
        return f"({n})"
    
    current = n
    result = ""
    divisor = 2
    
    while current > 1:
        count = 0
        while current % divisor == 0:
            current //= divisor
            count += 1
        
        if count > 0:
            if count == 1:
                result += f"({divisor})"
            else:
                result += f"({divisor}**{count})"
        
        divisor += 1
    
    return result

def test_razlo():
    assert razlo(1) == "(1)"
    assert razlo(0) == "(0)"

    assert razlo(13) == "(13)"
    
    assert razlo(8) == "(2**3)"      

    assert razlo(6) == "(2)(3)"    

    assert razlo(86240) == "(2**5)(5)(7**2)(11)"

    
    print("Все тесты пройдены")

test_razlo()

```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(n):
    current = 0
    k = 0
    
    while current < n:
        k += 1
        current += k**2
        
        if current == n:
            return k
    
    return 'It is impossible'

def test_pyramid():
    
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(30) == 4

    assert pyramid(2) == "It is impossible"
    assert pyramid(15) == "It is impossible"
    assert pyramid(23) == "It is impossible"

    assert pyramid(204) == 8

    assert pyramid(0) == "It is impossible"
    
    print("Все тесты пройдены")

test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def summa(n):
    s = 0
    for i in str(n):
        s += int(i)
    return s

def is_balanced(num):
    n = str(num)
    mid = len(n) // 2
    
    if len(n) % 2 == 0:
        left_part = n[:mid-1] 
        right_part = n[mid+1:] 
    else:
        left_part = n[:mid]  
        right_part = n[mid+1:] 
    
    return summa(left_part) == summa(right_part)


def test_is_balanced():
    assert is_balanced(1234) == False 
    assert is_balanced(1321) == True
    assert is_balanced(191282) == True 

    assert is_balanced(121) == True     
    assert is_balanced(12345) == False  

    assert is_balanced(11) == True
    assert is_balanced(1) == True

    print("Все тесты пройдены!")

    
test_is_balanced()
```
